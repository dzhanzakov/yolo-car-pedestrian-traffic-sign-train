```python
import cv2
from ultralytics import YOLO

model = YOLO("yolo11_large_cars_people/run_1/weights/best.pt")

# Open default camera
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Predict on current frame
    results = model.predict(frame, save=False, show=False)

    # Draw results on frame (optional)
    annotated_frame = results[0].plot()
    cv2.imshow("YOLO Prediction", annotated_frame)

    # Press any key to break
    if cv2.waitKey(1) != -1:
        print('key was pressed')
        break

# stop
cap.release()
cv2.destroyAllWindows()
```
